## USA500 PREDICTION

```python
import pandas as pd

# Load the datasets
file_paths = {
    "USA500": r"C:\Users\tthem\Downloads\USA500.IDXUSD_Candlestick_1_Hour_BID_01.01.2020-31.08.2024.csv",
    "USATECH": r"C:\Users\tthem\Downloads\USATECH.IDXUSD_Candlestick_1_Hour_BID_01.01.2020-31.08.2024.csv",
    "XAUUSD": r"C:\Users\tthem\Downloads\XAUUSD_Candlestick_1_Hour_BID_01.01.2020-31.08.2024.csv",
    "USA30": r"C:\Users\tthem\Downloads\USA30.IDXUSD_Candlestick_1_Hour_BID_01.01.2020-31.08.2024.csv",
    "BRENT": r"C:\Users\tthem\Downloads\BRENT.CMDUSD_Candlestick_1_Hour_BID_01.01.2020-31.08.2024.csv"
}

# Load all files into a dictionary of DataFrames
dataframes = {key: pd.read_csv(path) for key, path in file_paths.items()}
```

```python
# Converting the 'Local time' column to datetime for all datasets
for key, df in dataframes.items():
    dataframes[key]['Local time'] = pd.to_datetime(
        dataframes[key]['Local time'], format='%d.%m.%Y %H:%M:%S.%f GMT%z', utc=True
    )

# Merging the datasets on 'Local time', starting with USA500 as the target/base
merged_data = dataframes["USA500"].copy()
for key, df in dataframes.items():
    if key != "USA500":
        merged_data = pd.merge(
            merged_data, 
            df[['Local time', 'Close']], 
            on='Local time', 
            how='inner', 
            suffixes=('', f'_{key}')
        )
```

```python
# Defining the technical indicators

def calculate_sma(data, window):
    """Simple Moving Average (SMA)"""
    return data.rolling(window=window).mean()

def calculate_rsi(data, window=14):
    """Relative Strength Index (RSI)"""
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

def calculate_macd(data, short_window=12, long_window=26, signal_window=9):
    """MACD and Signal Line"""
    short_ema = data.ewm(span=short_window, adjust=False).mean()
    long_ema = data.ewm(span=long_window, adjust=False).mean()
    macd = short_ema - long_ema
    signal = macd.ewm(span=signal_window, adjust=False).mean()
    return macd - signal

# Add technical indicators for each instrument
for col in ['Close', 'Close_XAUUSD', 'Close_USATECH', 'Close_USA30', 'Close_BRENT']:
    prefix = "" if col == 'Close' else col.split('_')[1] + "_"
    merged_data[f'{prefix}SMA_10'] = calculate_sma(merged_data[col], window=10)
    merged_data[f'{prefix}SMA_50'] = calculate_sma(merged_data[col], window=50)
    merged_data[f'{prefix}RSI'] = calculate_rsi(merged_data[col])
    merged_data[f'{prefix}MACD_diff'] = calculate_macd(merged_data[col])

# Drop rows with NaN values resulting from technical indicator calculations
merged_data_cleaned = merged_data.dropna().reset_index(drop=True)

print(merged_data_cleaned.head())
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Initializing the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Trainning the model
rf_model.fit(X_train, y_train)

# Predicti9ng on the test set
y_pred = rf_model.predict(X_test)

# Evaluating the model
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

mse, mae, r2
```
