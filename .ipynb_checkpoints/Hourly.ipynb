```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor 
from sklearn.model_selection import train_test_split, TimeSeriesSplit
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor 
from sklearn.model_selection import train_test_split, TimeSeriesSplit
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Step 2: Data Collection and Preparation
# Load the dataset
data = pd.read_csv("USDJPY_Candlestick_1_Hour_BID_01.01.2020-31.08.2024.csv") 

# Convert 'Local time' to datetime format (Corrected format)
data['Local time'] = pd.to_datetime(data['Local time'], format='%d.%m.%Y %H:%M:%S.%f GMT%z') 
data.set_index('Local time', inplace=True)


# Step 3: Feature Engineering
# Calculate technical indicators (example with SMA and RSI)
data['SMA_20'] = data['Close'].rolling(window=20).mean()
data['RSI_14'] = 100 - (100 / (1 + (data['Close'].diff(1).fillna(0).apply(lambda x: max(x, 0)).rolling(window=14).sum() / data['Close'].diff(1).fillna(0).apply(lambda x: abs(min(x, 0))).rolling(window=14).sum())))

# Create lagged features (example with 1-day lag)
data['Close_Lagged_1'] = data['Close'].shift(1)

# Remove rows with NaN values (introduced by indicators and lags)
data.dropna(inplace=True)

# Step 4: Prepare Data for Modeling
# Define target variable (here, we'll predict the closing price - regression)
X = data[['SMA_20', 'RSI_14', 'Close_Lagged_1']]  # Features
y = data['Close']  # Target variable

# Feature scaling
scaler = MinMaxScaler()
X = scaler.fit_transform(X)

# Train-test split (using TimeSeriesSplit for time-series data)
tscv = TimeSeriesSplit(n_splits=5) 
for train_index, test_index in tscv.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

# Step 5: Build and Train the Model
# Initialize the Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42) 

# Train the model
rf_model.fit(X_train, y_train)

# Step 6 & 7: Model Evaluation and Hyperparameter Tuning (Simplified)
# Make predictions on the test set
y_pred = rf_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}")

# (For more advanced tuning, use GridSearchCV or RandomizedSearchCV with TimeSeriesSplit)

# Step 8: Make Predictions and Visualize
# Plot predicted vs. actual prices
plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test, label='Actual')
plt.plot(y_test.index, y_pred, label='Predicted')
plt.xlabel("Time")
plt.ylabel("USD/JPY Price")
plt.title("Actual vs. Predicted USD/JPY Price")
plt.legend()
plt.show()

```
